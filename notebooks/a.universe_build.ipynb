# Stock Universe Build

**Context**: "What stocks do you include in your strategy/alpha?"
Sometimes the alpha or strategy can not only focus on big caps stock 
but also select specified narrowed stocks and even penny stocks.
But the Survivorship Bias is a big problem you should not ignore,
and the fix is by introducing historical data.

**Goal**: Try to build a stock universe helper funtion as data entry.
Balance between boardness and Survivorship Bias problem when using it.
Given any date, trace back to the stocks fit the conditions as of date.
the function must not use information from after as_of_date.

---

## Attributes

1. **Country** — where is this company?
2. **Min Average Volume** — volume filter
3. **Exclude OTC** — bool to exclude otc market stocks
